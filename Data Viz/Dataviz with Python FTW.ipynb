## Import Datasets

We are going to us the [Iris](https://archive.ics.uci.edu/ml/datasets/iris) and the [Wine Reviews](https://www.kaggle.com/zynicide/wine-reviews) dataset.

```python
%matplotlib inline  
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('/content/iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
iris.head()
```

```python
wine_reviews = pd.read_csv('/content/winemag-data-130k-v2.csv', index_col=0)
wine_reviews.head(1)
```

## Matplotlib

Matplotlib is a Python plotting libary.

```python
import matplotlib.pyplot as plt
```

### Scatter plot

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(iris['sepal_length'], iris['sepal_width'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

We can color each point by class

```python
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
fig, ax = plt.subplots()
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],color=colors[iris['class'][i]])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

### Heatmap

Heatmaps are perfect for exploring the correlation of the features

```python
corr = iris.corr()
fig, ax = plt.subplots()
im = ax.imshow(corr.values)
```

```python
# get correlation matrix
corr = iris.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
```

```python
# get correlation matrix
corr = iris.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
```

```python
#Importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.__version__
from matplotlib import rcParams, cycler
```

```python
#Read excel file
indicators = pd.read_excel('/content/Data_Extract_From_World_Development_Indicators.xlsx', index_col=0)
indicators
```

```python
#Create df for People using at least basic sanitation services (% of population)
sanitation = indicators.loc[indicators['Series Name'] == 'People using at least basic sanitation services (% of population)']
sanitation
```

```python
#Create df for 2010 and 2015 values
sanitation_2010_2015 = sanitation[['2010 [YR2010]','2011 [YR2011]','2012 [YR2012]','2013 [YR2013]','2014 [YR2014]','2015 [YR2015]']]
sanitation_2010_2015
```

```python
#Create scatterplot
plt.rcdefaults()
rcParams['legend.loc'] = 'lower right'
rcParams['legend.fontsize'] = 'xx-small'
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False

sns.scatterplot(x='2010 [YR2010]', y='2015 [YR2015]', hue='Country Name', data=sanitation_2010_2015, s=100).set(title='2010 vs 2015')
```

We can explore if there is a positive linear correlation between 2010 and 2015

```python
#Create line graph of transposed sanitation_2010_2015
rcParams['xtick.labelsize'] = 'x-small'
rcParams['ytick.labelsize'] = 'large'

sns.lineplot(data=sanitation_2010_2015.T).set(title='People using at least basic sanitation services (% of population)')
```

Singapore consistently 100%
Increase for Vietnam, Philippines, Indonesia, Lao, and Cambodia
Decrease for Myanmar

```python
#Create df for Population, total
total_pop = indicators.loc[indicators['Series Name'] == 'Population, total']

#Create histogram
plt.rcdefaults()
sns.distplot(total_pop['2015 [YR2015]'], kde=False).set(title='Population distribution in 2015')
```

```python
#Create df for PH indicators
ph_indicators = indicators.loc[indicators['Country Code'] == 'PHL']

#Create bar graph
sns.countplot(ph_indicators['2020 [YR2020]']).set(title='PH Indicators in 2020')
```

```python
#Create stripplot

sns.stripplot(y='2015 [YR2015]', data=sanitation_2010_2015).set(title='People using at least basic sanitation services (% of population) in 2015')
```
